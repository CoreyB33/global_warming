```python
from csv import reader, writer 
with open('../data/global_emissions.csv') as f, open('../data/glob_emiss.csv', 'w') as fw: 
    writer(fw, delimiter=',').writerows(zip(*reader(f, delimiter=',')))
```

```python
ge = pd.read_csv('../data/glob_emiss.csv')
```

```python
ge.head()
```

```python
ge.tail()
```

```python
ge = ge.drop([0,1,2,58,59,60,61,62])
```

```python
ge.head()
```

```python
ge = ge.rename({'Country Name':'year'}, axis=1)
```

```python
ge.head()
```

```python
ge['R']
```

```python
cols = ge.columns
```

```python
ge[cols] = ge[cols].apply(pd.to_numeric, errors='coerce')
```

```python
sns.lineplot(ge.year,ge['China'])
sns.lineplot(ge.year,ge['India'])
sns.lineplot(ge.year,ge['United States'])
sns.lineplot(ge.year,ge['Russian Federation'])
sns.set(rc={'figure.figsize':(16,10)})
plt.xticks([1960,1970,1980,1990,2000,2010])
plt.yticks([0,1,2,3,4,5])
plt.ylabel('CO2 emissions (metric tons per capita)')
plt.legend(labels=['China','India','United States','Russia'])
```
